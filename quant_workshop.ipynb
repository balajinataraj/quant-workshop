### Price data from Centralized Exchange

```python
import ccxt
import datetime
```

```python


# Initialize the Binance exchange
exchange = ccxt.binance()

# Define the symbol for Bitcoin
symbol = 'BTC/USDT'

# Define the time frame for the last month
end_time = exchange.milliseconds()
start_time = end_time - 30 * 24 * 60 * 60 * 1000  # 30 days in milliseconds

# Fetch OHLCV (Open, High, Low, Close, Volume) candlestick data for the specified time frame
ohlcvs = exchange.fetch_ohlcv(symbol, timeframe='1d', since=start_time, limit=None)

# Convert timestamps to human-readable dates
ohlcvs = [(datetime.datetime.fromtimestamp(timestamp / 1000), open, high, low, close, volume) for timestamp, open, high, low, close, volume in ohlcvs]
dat = []
# Print the OHLCV data for the last month
for ohlcv in ohlcvs:
    dat.append(f"Time: {ohlcv[0]}, Open: {ohlcv[1]}, High: {ohlcv[2]}, Low: {ohlcv[3]}, Close: {ohlcv[4]}, Volume: {ohlcv[5]}")
```

```python
import requests
```

```python
# CoinGecko API

def get_hist_price(token):
    cgk_url = f"https://api.coingecko.com/api/v3/coins/{token}/market_chart?vs_currency=usd&days=max"
    res = requests.get(cgk_url)
    if res.status_code == 200:
        return res.json()
    else:
        print(f'Cannot get price data. {res.text}')
        return None
```

```python
a = get_hist_price('bitcoin')
```

```python
from datetime import datetime, timedelta
import pandas as pd

start_time = datetime(2023, 1, 1)
end_time = datetime(2023, 10, 1)

prices = []
for dt in list(pd.date_range(start_time, end_time, freq='D')):
    prices.append(get_hist_price('bitcoin', dt.strftime("%Y-%m-%d"))
```
